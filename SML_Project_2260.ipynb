```python
import pandas as pd
d=pd.read_csv('/content/Wine_Dataset.csv')
print(d)
```

```python
d.isnull()
```

```python
y=d['Alcohol']
print(y)
```

```python
x=d.drop('Alcohol',axis=1)
print(x)
```

```python
dd = (x - x.min()) / (x.max() - x.min())
print(dd)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dd, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.svm import SVC
lr=SVC()
m=lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
yp=m.predict(X_test)
print(accuracy_score(y_test,yp))
```

```python
print(classification_report(y_test,yp))
```

```python
print(confusion_matrix(y_test,yp))
```

```python
from sklearn.linear_model import Perceptron
lr1=Perceptron()
m=lr1.fit(X_train,y_train)
```

```python
yp1=m.predict(X_test)
print(accuracy_score(y_test,yp1))
```

```python
print(confusion_matrix(y_test,yp1))
```

```python
print(classification_report(y_test,yp1))
```

```python
from sklearn.linear_model import LogisticRegression
lr2=LogisticRegression(max_iter=1000)
m=lr2.fit(X_train,y_train)
```

```python
yp2=m.predict(X_test)
print(accuracy_score(y_test,yp2))
```

```python
print(confusion_matrix(y_test,yp2))
```

```python
print(classification_report(y_test,yp2))
```
