```python
import pandas as pd
import matplotlib.pyplot as plt
ppl=['Kiran','Arun','Vijay','Varun']
age=[25,30,35,40]
height=[145,151,165,173]
wt=[45,55,65,75]
df=pd.DataFrame({'Name':ppl,'Age':age,'Height':height,'Weight':wt})
print(df)
```

```python
#Plot,Grid,Color
plt.figure(figsize=(8,6))
plt.scatter(df['Age'],df['Weight'],color='red',marker='o')
plt.grid(True)
plt.xlabel('Age(years)')
plt.ylabel('Weight(kg)')
plt.title('Scatter plot of Age vs Weight')
plt.show()
```

```python
#Bar
plt.figure(figsize=(8,6))
plt.plot(df['Name'],df['Height'],color='blue',marker='o')
plt.grid(True)
plt.xlabel('Name')
plt.ylabel('Height(cm)')
plt.title('Bar chart of individuals Name/Height')
plt.show()
```
